```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
def ensemble_error(e_base, n):
    """計算 ensemble 出錯率"""
    k_min = (n // 2) + 1  # 多數錯的起點
    return sum(binom.pmf(k, n, e_base) for k in range(k_min, n+1))
```

```python
# Curve: ε_base ∈ [0, 1]
e_base_values = np.linspace(0, 1, 401)
n_list = [11, 51, 201]
results = {n: [ensemble_error(e, n) for e in e_base_values] for n in n_list}
```

```python
plt.figure(figsize=(8, 6))
for n in n_list:
    plt.plot(e_base_values, results[n], label=f"n={n}")
plt.plot(e_base_values, e_base_values, "k--", label="Base Classifier Error Rate")
plt.xlabel("ε_base")
plt.ylabel("ε_ensemble")
plt.title("ε_base vs ε_ensemble")
plt.legend()
plt.grid(True)
plt.show()
```
